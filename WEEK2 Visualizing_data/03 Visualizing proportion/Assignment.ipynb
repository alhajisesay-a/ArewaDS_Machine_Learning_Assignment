For the puropse of this assignment, i will be using the birds dataset to construct pie chart, Donut chart and waffle chart. Insted of using excell to visualize this proportions, we use python to analyze the relationships.

First import the necessary modules and load the dataset to visualize proportions 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
birds_df = pd.read_csv('../data/birds.csv')
birds_df
```

## Pie Chart

From the dataframe we observed it contains numeric and categorical data. Let plot a pie chart by considering the ScientificName and Minimum Length.

```python
import matplotlib.pyplot as plt
# Plotting a pie chart for the scientific names of birds
plt.figure(figsize=(12, 10))
data = birds_df.groupby(['ScientificName'])['MinLength'].sum()
print(data)
labels = data.index
values = data.values
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Bird Species by Scientific Name')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

From the plot, we observed that the data point are very much clustered thus making it difficult to derive insite from the data. Therefore we need to consider certain values to make our plot visiable by introducing nlargest() parameter.

```python
import matplotlib.pyplot as plt
# Plotting a pie chart for the scientific names of birds
plt.figure(figsize=(12, 10))
data = birds_df.groupby(['ScientificName'])['MinLength'].sum().nlargest(20) # Get the top 20 scientific names by total minimum length
print(data)
labels = data.index
values = data.values
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Bird Species by Scientific Name')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

From the chart, we observe that the scientific names: Cygnus olor and Grus americana are the largest.

Now let consider the Genus category using the pie chart

```python
import matplotlib.pyplot as plt
# Plotting the number of birds by Genus
plt.figure(figsize=(12, 6))
data = birds_df['Genus'].value_counts().nlargest(15)  # Get the top 15 genera by count
print(data)
labels= data.index
values = data.values
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Bird Species by Genus')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

From the Genus category on the birds dataset, Setophaga is the largest followed by Calidris 

## Donut Chart

A more visually interesting pie chart is a donut chart, which is a pie chart with a hole in the middle. Let's look at our data using this method.

let take a look at the various category of birds Family:

```python
import matplotlib.pyplot as plt
# Plotting the number of birds by Family using Donut chart
plt.figure(figsize=(10, 12))
data = birds_df['Family'].value_counts().nlargest(15)  # Get the top 15 families by count 
print(data)
labels = data.index
values = data.values
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.5))
plt.title('Distribution of Bird Species by Family (Donut Chart)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

From the chart the largest family of birds is Anatidae followed by Parulidae
