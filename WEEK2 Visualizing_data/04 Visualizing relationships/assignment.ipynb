In order to analyse the bees and their honey production over a period of time that saw losses in the bee colony population overall, we first import the dataset. This can be achieve by importing the necessary liberies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset in to a dataframe
honey_df = pd.read_csv('../data/honey.csv')
# Display the first few rows of the dataframe
honey_df.head()
```

Let's investigate how the numcol varies over the years across the states

```python
sns.relplot(data=honey_df,x='year',y='numcol',col='state',col_wrap=4,kind='line')
```

From the charts above the information is very much congested when the col_wrap is set to 4. Let adjust col_wrap parameter to 3 and observe the outcome.

```python
sns.relplot(data=honey_df,x='year',y='numcol',col='state',col_wrap=3,kind='line')
```

From the plots, we observed  in some plot that the number of colonies of bees decrease over time, at some point, it increases with time. And finally some maintain constant number of colonies over time.

Let ous consider another parameters such as the total produced per year in every number of colony.

```python
sns.relplot(data=honey_df,x='numcol',y='totalprod',col='year',col_wrap=3,kind='line')
```

From the plot it's is clear that the data points follows a regular patterns at the initial stage every year. However in some years the number of produce tend to fluctate.
