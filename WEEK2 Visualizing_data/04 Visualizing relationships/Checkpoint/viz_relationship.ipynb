# ðŸ”— Visualizing Relationships with Matplotlib
Visualizing relationships between variables helps uncover trends, patterns, and potential correlations.

Visualizing relationships can be done using various types of plots, including:

- Scatter Plots

- Line Charts

- Bubble Charts

- Pair Plots (via Seaborn)


### Scatter Plot ðŸŽ¯


A scatter plot shows how two numeric variables relate. Each point represents an observation, with the x and y coordinates showing its values.

- Ideal for correlation analysis

-  Not suitable for categorical comparisons

Use Cases:


- Height vs Weight analysis

- Age vs Income

- Advertising budget vs Sales

#### Example 1: Hours studied vs Exam score

```python
import matplotlib.pyplot as plt
import numpy as np

hours = np.array([1, 2, 3, 4, 5, 6, 7, 8])
scores = np.array([50, 55, 60, 65, 70, 75, 80, 85])

plt.figure(figsize=(6,4))
plt.scatter(hours, scores, color='teal')
plt.title("Hours Studied vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.grid(True)
plt.show()
```

#### Example 2: Engine size vs Fuel efficiency

```python
engine_size = [1.0, 1.5, 1.6, 2.0, 2.5, 3.0, 3.5]
mpg = [40, 38, 35, 30, 28, 22, 20]

plt.figure(figsize=(6,4))
plt.scatter(engine_size, mpg, color='orange')
plt.title("Engine Size vs MPG")
plt.xlabel("Engine Size (L)")
plt.ylabel("Miles per Gallon")
plt.grid(True)
plt.show()
```

### Line Chart ðŸ“ˆ


A line chart connects data points in a sequenceâ€”often used for time series or continuous data.


- Best for trends over time

- Doesnâ€™t show distributions or clusters

Use Cases:

- Stock prices over time

- Daily temperatures

- Sales over months

#### Example 1: Monthly revenue trend

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
revenue = [1000, 1200, 1400, 1600, 1550, 1650]

plt.figure(figsize=(6,4))
plt.plot(months, revenue, marker='o', color='navy')
plt.title("Monthly Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue ($)")
plt.grid(True)
plt.show()
```

#### Example 2: Website visits over 7 days

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
visits = [120, 135, 150, 160, 155, 200, 180]

plt.plot(days, visits, marker='s', linestyle='--', color='green')
plt.title("Website Traffic Over a Week")
plt.xlabel("Day")
plt.ylabel("Visitors")
plt.grid(True)
plt.show()
```

####  Bubble Chart

A bubble chart is a scatter plot where each point also has a size (3rd dimension), showing a third variable visually.


- Great for showing impact with size
- Can get cluttered with many points



Use Cases:

- Product sales vs profit vs volume

- Population vs area vs GDP

- Social media engagement

Example 1: Marketing budget vs sales vs conversion

```python
budget = [100, 200, 300, 400, 500]
sales = [10, 25, 30, 40, 55]
conversions = [20, 30, 40, 60, 80]  # bubble size

plt.scatter(budget, sales, s=conversions, alpha=0.5, c='purple')
plt.title("Marketing Spend vs Sales (Bubble = Conversions)")
plt.xlabel("Marketing Budget ($k)")
plt.ylabel("Sales ($k)")
plt.grid(True)
plt.show()
```

Example 2: GDP vs Population vs Area

```python
countries = ['A', 'B', 'C', 'D']
gdp = [1.5, 2.0, 2.5, 3.5]
population = [30, 40, 50, 60]
area = [100, 150, 300, 200]

plt.scatter(gdp, population, s=area, alpha=0.6, color='skyblue')
plt.title("GDP vs Population (Bubble = Area)")
plt.xlabel("GDP (Trillions)")
plt.ylabel("Population (Millions)")
plt.grid(True)
plt.show()
```

## Facet grids

- A FacetGrid lets you create a matrix of plots based on the values of one or more categorical variables. Each subplot is a subset (facet) of the original dataset.


- Great for comparison across categories

- Not ideal for small datasets or rare categories


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset('titanic')
titanic
```

```python
# Create the FacetGrid
g = sns.FacetGrid(titanic, row='sex', col='class', margin_titles=True)

# Map a scatterplot to the grid
g.map(sns.scatterplot, 'age', 'fare', alpha=0.6)

# Add titles and show
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle("Titanic: Age vs Fare by Gender and Class")
plt.show()
```

The plot shows: 

- Horizontal panels = passenger class (class)

- Vertical panels = passenger gender (sex)

- Each scatterplot shows how age relates to fare paid, split by these subgroup

#### Pair Plot (via Seaborn)

A pair plot shows scatter plots between all numeric pairs in a dataset and histograms on the diagonal. Ideal for quick exploratory data analysis (EDA).

- Powerful for multivariate data

- Requires clean, numeric dataset


#### Example 1: Iris dataset


```python
import seaborn as sns
df = sns.load_dataset('iris')
df
```

```python

sns.pairplot(df, hue='species')
plt.suptitle("Iris Pair Plot", y=1.02)
plt.show()
```
