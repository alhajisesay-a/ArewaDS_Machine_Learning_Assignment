From the dateframe, we observe the dataset contains a mixed of numeric and categoric data.
Let explor the data by using a line plot to visualise the relations between maxwingspan.

```python
maxwingspan = birds['MaxWingspan']
wingspan = maxwingspan.max()
print(f"The maximum wingspan is {wingspan} cm.")
# Plot the distribution of the MaxWingspan
maxwingspan.plot()#color='blue', linestyle='--', marker='o')
plt.title('Distribution of Max Wingspan')
plt.xlabel('Max Wingspan (cm)')
plt.ylabel('Wingspan (cm)')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

From the graph, we notice the presence of outliers from two data point and the maximum wingspan is 2300 cm
Let find out which birds with wingspan greater than 400cm from the figure above

```python
# plot the data
plt.figure(figsize=(10, 6))
plt.title('Distribution of Max Wingspan')
plt.xlabel('Max Wingspan (cm)')
plt.ylabel('Wingspan (cm)')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)
plt.grid(axis='y', alpha=0.75)
for i in range(len(birds)):
    x = birds['Name'][i]
    y = birds['MaxWingspan'][i]
    plt.plot(x,y, marker='o', markersize=8, color='blue', alpha=0.6)
    if birds['MaxWingspan'][i] > 350:
        plt.text(x, y - 0.10, birds['Name'][i], fontsize=12)

plt.show()
```

From the line plot, it is crystal clear that the name of birds that has maxwingspan greater that 350cm are Prairie falcon and Bald eagle and as such, account for outliers on the dataset

Next step is to clean the dataset by taking care of outliers.

```python
# plot the data
plt.figure(figsize=(10, 6))
plt.title('Distribution of Max Wingspan')
plt.xlabel('Max Wingspan (cm)')
plt.ylabel('Wingspan (cm)')
plt.tick_params(axis='both', which='both', labelbottom=False, bottom=False)
plt.grid(axis='y', alpha=0.75)
for i in range(len(birds)):
    x = birds['Name'][i]
    y = birds['MaxWingspan'][i]
    #plt.plot(x, y, marker='o', markersize=8, color='blue', alpha=0.6)
    if birds['MaxWingspan'][i] < 350:
        plt.plot(x,y, marker='o', markersize=8, color='b', alpha=0.6)
plt.show()
```

Now we visualize the relationship between MinLength  and MinBodyMass using scatterplot and add a third parameter called Category. The scatter plot shows the relationship between two numeric variables in a detaset.

```python
import matplotlib.pyplot as plt
# Create a scatter plot for MinLength vs MinBodyMass
plt.figure(figsize=(10, 6))
plt.scatter(birds['MinLength'], birds['MinBodyMass'], c=birds['Order'].astype('category').cat.codes, cmap='viridis', alpha=0.6)
plt.title('Scatter Plot of MinLength vs MinBodyMass')
plt.xlabel('Min Length (cm)')
plt.ylabel('Min Body Mass (g)')
plt.show()
```

From the scatter plot, we observed patterns on the dataset. It shows that there is a steady increase in minimum bodys mass from minimum lenght between 3cm to 29cm. Also we can conclude that the as the minimum length of the birds increase so as the minimum body mass in grams.

Now that we have visualized the relationship between minimum body mass and that of minimum length using scatter plot, let dive deeper in the dataset to obtain more information using groupby category to compaire minmum length and maximum length

```python
birds['Order'].value_counts().plot(kind='barh',figsize=(12, 12), color='skyblue')
plt.title('Number of Birds per Category')
plt.xlabel('Number of Birds')
plt.ylabel('Order')
plt.show()
```

This bar chart shows a good view of the number of birds in each Order. In a blink of an eye, you see that the largest number of birds in this region are in the Passeriformes order. Coraciiformes is the least!

```python
import matplotlib.pyplot as plt
# Create a horizontal bar plot for MinLength and MaxLength by Category
plt.figure(figsize=(12, 12))
minLength = birds['MinLength']
maxLength = birds['MaxLength']
category = birds['Order']

plt.barh(category, maxLength)
plt.barh(category, minLength)
plt.title('MinLength and MaxLength by Order')
plt.xlabel('Length (cm)')
plt.ylabel('Category')
plt.legend(['Max Length', 'Min Length'])
plt.grid(axis='x', alpha=0.75)

plt.show()
```
